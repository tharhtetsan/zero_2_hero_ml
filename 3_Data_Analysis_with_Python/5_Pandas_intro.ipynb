### Data frame index call and change

```python
import pandas as pd

data = {
  "calories": [420, 380, 390],
  "duration": [50, 40, 45]
}

data_df = pd.DataFrame(data)
print(data_df)
print(data_df["calories"])
```

```python
cal0,dur0 = (data_df.loc[0])
print(cal0)
print(dur0)
```

```python
data1_df = pd.DataFrame(data,index = [2,4,6])
print(data1_df)
```

```python
cal0,dur0 = (data1_df.loc[6])
print(cal0)
print(dur0)
```

```python
test = (data1_df["calories"])
print(test)
data1_df.to_csv("test.csv")
```

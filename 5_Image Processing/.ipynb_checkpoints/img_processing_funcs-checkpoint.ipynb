```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img_path = r"test_img/test_3.png"
img = cv2.imread(img_path)
rgb_img = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
print(img.shape)
plt.imshow(rgb_img)
```

```python
img_gray = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
print(img_gray.shape)
plt.imshow(img_gray)
```

```python
gaussianBlur_img = cv2.GaussianBlur(img_gray,(5,5),0)
print(gaussianBlur_img.shape)
plt.imshow(gaussianBlur_img)
```

```python
thres_hold_img = cv2.threshold(gaussianBlur_img,100,255, cv2.THRESH_BINARY)[1]
print(thres_hold_img.shape)

plt.imshow(thres_hold_img)
```

```python
gray = cv2.bilateralFilter(thres_hold_img, 1, 100, 150)
plt.imshow(gray)
```

```python
imgThreshold = cv2.Canny(gray, 20, 250)
plt.imshow(imgThreshold)
```

```python
kernel = cv2.getStructuringElement(cv2.MORPH_RECT, (3, 3))
closed = cv2.morphologyEx(imgThreshold, cv2.MORPH_CLOSE, kernel)
```

```python
plt.imshow(closed)
```

```python
plt.imshow(kernel)
```

```python
contours, hierarchy = cv2.findContours(closed, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
```

```python
def reorder(myPoints):
    myPoints = myPoints.reshape((4, 2))
    myPointsNew = np.zeros((4, 1, 2), dtype=np.int32)
    add = myPoints.sum(1)

    myPointsNew[0] = myPoints[np.argmin(add)]
    myPointsNew[3] = myPoints[np.argmax(add)]
    diff = np.diff(myPoints, axis=1)
    myPointsNew[1] = myPoints[np.argmin(diff)]
    myPointsNew[2] = myPoints[np.argmax(diff)]

    return myPointsNew
```

```python
rect_list = []
for i in contours:
    area = cv2.contourArea(i)

    if area > 8000:
        peri = cv2.arcLength(i, True)
        approx = cv2.approxPolyDP(i, 0.02 * peri, True)
        if len(approx) == 4:

            #cv2.drawContours(rgb, [approx], -1, (0, 255, 0), 2)
            data = approx
            rect_list.append(reorder(approx))
            # cv2.drawContours(img, [approx], -1, (0, 255, 0),2)
        elif len(approx) > 4:
            #cv2.drawContours(rgb, [approx], -1, (0, 255, 0), 2)
            data = approx
            rect_list.append(reorder(approx[:4]))
            # cv2.drawContours(img, [approx], -1, (0, 255, 0),2)
```

```python
rect_list =  rect_list[0]
```

```python
print(rect_list)
start_point = (rect_list[0][0][0],rect_list[0][0][1])
end_point = (rect_list[1][0][0],rect_list[1][0][1])

print(start_point)
print(end_point)
```

```python
color = (255, 0, 0)
thickness = 2
image = cv2.rectangle(rgb_img, start_point, end_point, color, thickness)
plt.imshow(image)
```

### Circle detection

```python
img_path = r"test_img/test_1.png"
img = cv2.imread(img_path)
rgb_img = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
gray = cv2.cvtColor(rgb_img, cv2.COLOR_BGR2GRAY) 
gray_blurred = cv2.blur(gray, (3, 3))
plt.imshow(gray_blurred)
```

```python
detected_circles = cv2.HoughCircles(gray_blurred,  
                   cv2.HOUGH_GRADIENT,dp = 1, minDist  = 2, param1 = 50, 
               param2 = 50, minRadius = 1, maxRadius = 40) 
print(detected_circles)
```

```python
if detected_circles is not None: 
    detected_circles = np.uint16(np.around(detected_circles)) 
    for pt in detected_circles[0, :]:
        a, b, r = pt[0], pt[1], pt[2] 
        cv2.circle(rgb_img, (a, b), r, (0, 255, 0), 2)
plt.imshow(rgb_img)
```

```python
img = cv2.imread(r"test_img/test_3.png")
img = cv2.resize(img, (500, 400))
print(img.shape)
output = img.copy()
gray_img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
blur = cv2.medianBlur(gray_img, 5)
circles = cv2.HoughCircles(blur, cv2.HOUGH_GRADIENT,
                          1, 100, 10, param1=130, param2=55, minRadius=5, maxRadius=0)

detected_circles = np.uint16(np.around(circles))
print(detected_circles)
for pt in detected_circles[0, :]:
    x, y ,r = pt[0], pt[1], pt[2] 
    cv2.circle(output, (x, y), r, (0, 255, 0), 3)
    cv2.circle(output, (x, y), 2, (255, 0, 0), 3)
plt.imshow(output)
```
